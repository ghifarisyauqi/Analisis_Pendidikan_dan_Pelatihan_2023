## **Data Understanding and Cleaning**
Sebelum masuk ke dalam analisis, kita perlu mengenal dataset kita lebih jauh dalam tahapan *data understanding*.
```python
df.info()
```

- Dataset berisi 215 baris dan 26 kolom dengan tipe object secara keseluruhan
- Terdapat beberapa missing value pada beberapa kolom. Hanya saja missing value pada dataset ini berarti nilainya adalah 0
- Kita akan mengganti beberapa tipe data sesuai dengan seharusnya.

### **Mengatasi Missing Value**

```python
# Mengisi missing value pada beberapa kolom dengan nilai 0
df['Konsumsi'] = df['Konsumsi'].fillna(0)
df['Honor_Penyelenggara'] = df['Honor_Penyelenggara'].fillna(0)
df['Honor_Pengajar'] = df['Honor_Pengajar'].fillna(0)
df['Jasa_Lainnya'] = df['Jasa_Lainnya'].fillna(0)
df['PIC_Penyelenggara'] = df['PIC_Penyelenggara'].fillna('0')
df['Supporting'] = df['Supporting'].fillna('0')
df['PIC_Akademis'] = df['PIC_Akademis'].fillna('0')
df['PIC_NPSK'] = df['PIC_NPSK'].fillna('0')
df['Widyaiswara'] = df['Widyaiswara'].fillna('0')
df['PIC_Sarana_Prasarana_IT'] = df['PIC_Sarana_Prasarana_IT'].fillna('0')
df['SPPD_Panitia'] = df['SPPD_Panitia'].fillna(0)
df['SPPD_Peserta'] = df['SPPD_Peserta'].fillna(0)
df['SPPD_Pengajar'] = df['SPPD_Pengajar'].fillna(0)
```

- Pada proses Cleaning, ditemukan ada perbedaan format nilai pada kolom `SPPD_Panitia`, `SPPD_Pengajar`, dan `SPPD_Peserta` sehingga kita akan mengubah nilainya menjadi satu format yang sama

```python
df.loc[df['SPPD_Panitia'] == 'Rp.17.480.000', 'SPPD_Panitia'] = 17480000
df.loc[df['SPPD_Panitia'] == '-', 'SPPD_Panitia'] = 0
df.loc[df['SPPD_Peserta'] == 'Rp.63.130.000', 'SPPD_Peserta'] = 63130000
df.loc[df['SPPD_Peserta'] == 'Rp.23.290.000', 'SPPD_Peserta'] = 23290000
df.loc[df['SPPD_Pengajar'] == 'Rp.15.000.000', 'SPPD_Pengajar'] = 15000000
df.loc[df['SPPD_Pengajar'] == '-', 'SPPD_Pengajar'] = 0
```

**Mengisi missing value pada kolom `Tempat_Pelaksanaan`**

```python
# Dataset yang terdapat missing value pada kolom Tempat_Pelaksanaan
df[df['Tempat_Pelaksanaan'].isna()]
```

- Kita akan mengisi missing value di atas dengan nilai yang sering muncul pada Tempat_Pelaksanaan apabila pelatihan diselenggarakan di Jakarta

```python
df[df['Kota_Pelaksanaan']=='Jakarta'][['Tempat_Pelaksanaan']].value_counts()
```

- Apabila pelatihan diselenggarakan di Jakarta, BPSDM menjadi tempat pelaksanaan terbanyak sehingga kita akan mengisi missing value dengan BPSDM

```python
df['Tempat_Pelaksanaan'] = df['Tempat_Pelaksanaan'].fillna('BPSDM')
```
